# Here you can easily see some of the optimal trajectories of the database.

Imports

```python
from matplotlib import pyplot as plt
import glob
import os
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

Get the current path

```python
path = os.path.abspath(os.getcwd())
```

Get all the files in the database directory (here called /opt_traj)

```python
files = glob.glob(path+'/opt_traj/'+'*.dat') # Get all .dat files in the folder
```

Let's plot a trajectory

```python
file_number = 0
fig, axs = plt.subplots(2,3,figsize=(11,7), tight_layout=True)


for file_number in range(len(files)):
    data = np.loadtxt(files[file_number])
    # matplotlib.rcParams['text.usetex'] = True # habilitando tex 

    # # Get the q value 
    # start = files[file_number].find('q_') + 2
    # end = files[file_number].find('r_', start)
    # q_value = files[file_number][start:end]

    # # get the r value 
    # start = files[file_number].find('r_') + 2
    # end = files[file_number].find('x1i_', start)
    # r_value = files[file_number][start:end]
    

    axs[0,0].plot(data[:,0],data[:,1])
    axs[0,0].set_xlabel(r'$s$')
    axs[0,0].set_ylabel(r'$[m]$')

    axs[0,1].plot(data[:,0],data[:,2])
    axs[0,1].set_xlabel(r'$s$')
    axs[0,1].set_ylabel(r'$[m/s]$')

    axs[1,0].plot(data[:,0],data[:,3]*180/np.pi)
    axs[1,0].set_xlabel(r'$s$')
    axs[1,0].set_ylabel(r'$[^\circ]$')

    axs[1,1].plot(data[:,0],data[:,4]*180/np.pi)
    axs[1,1].set_xlabel(r'$s$')
    axs[1,1].set_ylabel(r'$[^\circ/s]$')

    axs[0,2].plot(data[:,0],data[:,5])
    axs[0,2].set_xlabel(r'$s$')
    axs[0,2].set_ylabel(r'$[N]$')

    # axs[1,2].plot([0],[0], label = 'Trajetria nmero: ' + str(file_number))

    axs[1,2].legend(loc='upper left')
```

```python
plt.figure(figsize=(11,7))
plt.plot(data[:,0],data[:,5])
plt.grid(True)
plt.xlabel(r'$s$')
plt.ylabel(r'$[N]$')
```
